# **Why Use Databases?**

Databases help us:
- **Organize Information**: Keep everything neat and tidy.
- **Find Information Quickly**: Instead of searching through piles of paper, we can search through the database.
- **Update Information Easily**: If something changes (like your favorite movie), you can change it in one place, and it's updated everywhere.

### **Key Components of a Database**

1. **Tables**: The main part of a database, where data is stored. Each table has rows and columns.
   - **Rows**: Each row represents a single record (like one book).
   - **Columns**: Each column holds a specific type of information (like title, author, and year published).

2. **Records**: Each row in a table is a record that contains related information.

3. **Fields**: Each column in a table is called a field, which holds specific data about the record.

# **What is a Database Management System (DBMS)?**

- **Definition**: A **DBMS** is software that allows users to create, manage, and manipulate databases. It provides tools for interacting with the database, making it easy to retrieve, update, and organize data.

- **Example**: Examples of popular DBMS software include:
  - **MySQL**: Often used for web applications.
  - **Microsoft Access**: A user-friendly option for small databases.
  - **Oracle**: Used for larger, enterprise-level applications.

- **Key Characteristics**:
  - Acts as a bridge between users and the database.
  - Provides functionalities such as data entry, querying, and reporting.
  - Ensures data security and integrity.

### **Key Differences Between Database and DBMS**

| Feature                      | Database                                      | DBMS                                      |
|------------------------------|-----------------------------------------------|-------------------------------------------|
| **Purpose**                  | Stores data in an organized way               | Manages and manipulates the data stored in a database |
| **Functionality**            | Just holds the data                           | Provides tools for creating, accessing, and managing databases |
| **Examples**                 | A collection of records, like a book list    | Software like MySQL, Microsoft Access, Oracle |
| **User Interaction**         | Users access data through a DBMS             | Users interact with the DBMS to perform operations on the database |
| **Data Management**          | No built-in management features                | Includes features for data manipulation, retrieval, and security |


# **1. Introduction to SQL for Data Analytics**

**SQL** (Structured Query Language) is a domain-specific programming language designed to handle and manipulate data in tables. Developed in the 1970s, SQL is still widely used across various systems that manage business dataâ€”whether they store data in traditional tables or alternative structures.

##**1.1 What Does SQL Stand For?**

#### **S is for Structured**
SQL is a language used to work with data structured as tables. Examples include:
- **Excel** spreadsheets containing financial data
- **Google BigQuery** tables with census statistics
- **Relational Database Management Systems (RDBMS)**, like MySQL and PostgreSQL, used by online stores to handle customer, product, and order data.

Tables in SQL are organized into **rows** and **columns**:
- Each **row** represents an individual object or entity (like a customer or order).
- Each **column** represents a property or attribute (like name or birth year).

#### **Q is for Query**
SQL is a **declarative language**, meaning you tell the system *what* you want, not *how* to do it. For instance, to retrieve all rows and columns from the census table:

```sql
SELECT * FROM census;
```

- `*` selects all columns.
- Ending a SQL statement with a **semicolon (;)** is a good practice to avoid errors.
- SQL **keywords** (like `SELECT`, `WHERE`) are not case-sensitive, but theyâ€™re often written in uppercase for readability.

#### **L is for Language**
SQL reads like a natural language, making it relatively easy to understand:
- For example, "SELECT everything from the census" describes exactly what `SELECT * FROM census;` does.
- SQL was standardized in the 1980s, but different databases have slight variations, or **dialects**. For example, MySQL and SQL Server may handle dates differently, but knowing basic SQL makes it easy to adapt.

### 1. Data Definition Language (DDL)

- **Purpose**: To define and manage database structures (like tables).
- **Key Commands**: `CREATE`, `ALTER`, `DROP`
- **Example**: Creating an Orders table.
  ```sql
  CREATE TABLE Orders (
      order_id INT PRIMARY KEY,
      customer_id INT,
      order_date DATE,
      total_amount DECIMAL(10, 2),
      status VARCHAR(20)
  );
  ```

---

### 2. Data Query Language (DQL)

- **Purpose**: To retrieve data from the database.
- **Key Command**: `SELECT`
- **Example**: Getting all orders.
  ```sql
  SELECT * FROM Orders;
  ```

---

### 3. Data Manipulation Language (DML)

- **Purpose**: To manipulate data in the database (add, update, delete).
- **Key Commands**: `INSERT`, `UPDATE`, `DELETE`
- **Example**: Inserting a new order.
  ```sql
  INSERT INTO Orders (order_id, customer_id, order_date, total_amount, status)
  VALUES (1, 101, '2024-10-25', 150.75, 'Shipped');
  ```

---

### 4. Data Control Language (DCL)

- **Purpose**: To control access to data (permissions).
- **Key Commands**: `GRANT`, `REVOKE`
- **Example**: Granting permission to a user.
  ```sql
  GRANT SELECT ON Orders TO user_name;
  ```

---

### 5. Transaction Control Language (TCL)

- **Purpose**: To manage transactions to ensure data integrity.
- **Key Commands**: `COMMIT`, `ROLLBACK`
- **Example**: Committing a transaction.
  ```sql
  BEGIN;

  INSERT INTO Orders (order_id, customer_id, order_date, total_amount, status)
  VALUES (2, 102, '2024-10-26', 200.00, 'Processing');

  COMMIT;  -- Saves the changes
  ```

---

### Summary

- **DDL**: Manages **structure** (tables, schemas).
- **DQL**: Retrieves **data** (queries).
- **DML**: Manipulates **data** (insert, update, delete).
- **DCL**: Controls **permissions** (who can access what).
- **TCL**: Manages **transactions** (ensures changes are saved or reverted).

---

### Data Seekho Tips

- Use DDL to set up your database.
- Use DQL to fetch data as needed.
- Use DML to change data.
- Use DCL to manage user access.
- Use TCL to keep your data accurate during multiple operations.

### 1. **SELECT Statement**

The `SELECT` statement is used to retrieve data from a database.

**Syntax:**
```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition;
```

**Example:**
```sql
SELECT product_name, price
FROM products
WHERE category = 'Electronics';
```
*This retrieves the names and prices of all products in the "Electronics" category from the products table.*

---

### 2. **SELECT All Columns**

To select all columns from a table, use the asterisk (`*`).

**Syntax:**
```sql
SELECT *
FROM table_name;
```

**Example:**
```sql
SELECT *
FROM products;
```
*This retrieves all columns from the products table.*

---

### 3. **WHERE Clause**

The `WHERE` clause filters records based on specified conditions.

**Syntax:**
```sql
SELECT column1, column2, ...
FROM table_name
WHERE condition;
```

**Example:**
```sql
SELECT customer_name, order_date
FROM orders
WHERE status = 'Shipped';
```
*This retrieves the names of customers and order dates for all shipped orders.*

---

### 4. **ORDER BY Clause**

The `ORDER BY` clause is used to sort the result set.

**Syntax:**
```sql
SELECT column1, column2, ...
FROM table_name
ORDER BY column1 [ASC|DESC];
```

**Example:**
```sql
SELECT product_name, price
FROM products
ORDER BY price ASC;
```
*This retrieves product names and prices from the products table, sorted by price in ascending order.*

---

### 5. **GROUP BY Clause**

The `GROUP BY` clause groups rows that have the same values into summary rows.

**Syntax:**
```sql
SELECT column1, COUNT(*)
FROM table_name
GROUP BY column1;
```

**Example:**
```sql
SELECT category, COUNT(*)
FROM products
GROUP BY category;
```
*This retrieves the count of products grouped by category.*

---

### 6. **HAVING Clause**

The `HAVING` clause is used to filter groups based on aggregate functions.

**Syntax:**
```sql
SELECT column1, COUNT(*)
FROM table_name
GROUP BY column1
HAVING COUNT(*) > value;
```

**Example:**
```sql
SELECT category, AVG(price) AS average_price
FROM products
GROUP BY category
HAVING AVG(price) > 100;
```
*This retrieves categories where the average price of products is greater than $100.*

---

### 7. **JOIN Clauses**

`JOIN` clauses are used to combine rows from two or more tables based on a related column.

**Syntax:**
```sql
SELECT columns
FROM table1
JOIN table2 ON table1.common_column = table2.common_column;
```

**Example:**
```sql
SELECT c.customer_name, o.order_date
FROM customers c
JOIN orders o ON c.customer_id = o.customer_id;
```
*This retrieves customer names and their order dates by joining the customers and orders tables.*

---

### 8. **INSERT INTO Statement**

The `INSERT INTO` statement is used to add new records to a table.

**Syntax:**
```sql
INSERT INTO table_name (column1, column2, ...)
VALUES (value1, value2, ...);
```

**Example:**
```sql
INSERT INTO products (product_name, price, category)
VALUES ('Wireless Mouse', 25.99, 'Electronics');
```
*This adds a new product to the products table for a wireless mouse priced at $25.99.*

---

### 9. **UPDATE Statement**

The `UPDATE` statement is used to modify existing records in a table.

**Syntax:**
```sql
UPDATE table_name
SET column1 = value1, column2 = value2, ...
WHERE condition;
```

**Example:**
```sql
UPDATE products
SET price = 19.99
WHERE product_name = 'Wireless Mouse';
```
*This updates the price of the wireless mouse to $19.99 in the products table.*

---

### 10. **DELETE Statement**

The `DELETE` statement is used to remove records from a table.

**Syntax:**
```sql
DELETE FROM table_name
WHERE condition;
```

**Example:**
```sql
DELETE FROM products
WHERE product_name = 'Old Laptop';
```
*This deletes the product named "Old Laptop" from the products table.*

---

### 11. **Aggregate Functions**

SQL provides aggregate functions to perform calculations on a set of values.

- **COUNT**: Counts the number of rows.
- **SUM**: Calculates the total sum.
- **AVG**: Computes the average value.
- **MIN**: Finds the minimum value.
- **MAX**: Finds the maximum value.

**Example:**
```sql
SELECT SUM(total_amount) AS total_sales
FROM orders;
```
*This calculates the total sales from the orders table.*

---

### 12. **DISTINCT Keyword**

The `DISTINCT` keyword is used to return only unique values.

**Syntax:**
```sql
SELECT DISTINCT column1
FROM table_name;
```

**Example:**
```sql
SELECT DISTINCT category
FROM products;
```
*This retrieves a list of unique product categories from the products table.*
